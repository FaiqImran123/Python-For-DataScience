## Seaborn

### Types of Functions 
- Figure Function
- Axes Function 

### Main Classification
- Relational Graphs (scatter, lineplot)
- Dist plot (Histogram, kdeplot, rugplot)
- Catagorical (Barplot, boxplot, count plot, violin, swarmplot)
- Matrix Plot (heatmap, clusterplot)
- Multiplots (pair plot, joint plot) 


```python
import seaborn as sns
```

## Relational Graph 
- Scatter Plot
- Line plot

```python
# scatter plot
tips =sns.load_dataset("tips")
tips
```

```python
# scatter plot --> Axes function
sns.scatterplot(data =tips, x ="total_bill",y ="tip", hue ="sex", style="time", size ="size")
```

```python
# figure lvl function
sns.relplot(data =tips, x ="total_bill", y ="tip", kind ="scatter", hue="sex", style="time", size ="size")
```

```python
import plotly.express as px
```

```python

gap =px.data.gapminder()
gap
```

```python
tmp =gap[gap["country"]=="Pakistan"]
tmp
```

```python
# lineplot
sns.lineplot(data =tmp, x ="year", y ="lifeExp")
```

```python
sns.relplot(data =tmp, x ="year", y ="lifeExp", kind ="line")
```

```python
tmp =gap[gap["country"].isin(["Pakistan" , "India", "Brazil"])]
tmp
```

```python
sns.relplot(data =tmp, x ="year", y ="lifeExp", kind ="line", hue ="country", style ="continent", size ="continent")
```

```python
# facet plot
# col ="gender" --> Two differ col for gender
sns.relplot(data =tips, x ="total_bill", y ="tip", col ="smoker", row ="sex")
```

## Distribution Graph
- Univariate Analysis
- To Find Distrubution
- Plots --> Histogram, kdeplot, rugplot

```python
# axis level
sns.histplot(data =tips, x ="total_bill")
```

```python
# figure lvl
sns.displot(data =tips , x ="total_bill", kind ="hist")
```

```python
# bins
sns.displot(data =tips , x ="total_bill", kind ="hist", bins =50)
```

```python
tips
```

```python
# element =step --> To make it glassy look
sns.displot(data =tips , x ="total_bill", kind ="hist", hue ="sex", element ="step")
```

```python
# can make it side by side using col
sns.displot(data =tips , x ="total_bill", kind ="hist", col ="sex")
```

```python
sns.kdeplot(data =tips , x="total_bill", hue="sex")
```

```python
sns.rugplot(data =tips , x="total_bill", hue="sex")
```

## Matrix
- HeatMap
- ClusterMap

```python
gap
```

```python
grid =gap.pivot_table(index ="country", columns= "year", values="lifeExp", aggfunc="sum")
grid
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize =(10, 30))
# annot =true to get actual value
sns.heatmap(grid, annot=True, linewidth =0.5)
```

## Catagorical Plots

## Catagorical Scatter Plots

```python
tips
```

```python
# stripplot --> Scatter plot for one catagorical and numerical
sns.stripplot(data =tips, x ="day", y ="total_bill", jitter=False, hue ="sex")
```

```python
# figure level function
sns.catplot(data =tips, x ="day", y ="total_bill", kind ="strip", jitter =False)
```

```python
# swarm plot
sns.catplot(data =tips, x ="day", y ="total_bill", kind ="swarm", hue="sex")
```

## Catagorical Dist Plot

```python
# boxplot
sns.boxplot(data =tips, y ="total_bill")
```

```python
# boxplot
sns.boxplot(data =tips ,x ="sex",y ="total_bill", hue="day")
```

```python
# voilinplot --> kde + boxplot
sns.violinplot(data =tips,x ="sex", y ="total_bill")
```

```python
sns.kdeplot(data =tips, x ="total_bill", hue="sex")
```

```python
import pandas as pd
data =pd.DataFrame({"Product": [1,2,1,2], "Quarter": [1,1,2,2], "Sales":[10, 20, 30, 40]})
data
```

```python
g =data.groupby("Product")
```

```python
for group, data in g:
    q =data.groupby("Quarter").sum().reset_index()
    plt.scatter(q["Quarter"], q["Sales"])
```
