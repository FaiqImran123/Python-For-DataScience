## 2D Line Plot

```python
# Plotting a simple function
price =[10000, 30000, 80000, 70000, 50000]
year =[2008, 2009, 2010, 2011, 2012]
plt.plot(year, price)
plt.grid()
```

```python
# Reading csv file
df =pd.read_csv("datasets\sharma-kohli.csv")
df
```

```python
# Showing a V Kholi Scores Over the year
plt.plot(df["index"], df["V Kohli"])
```

```python
# Multiple Plotting
plt.plot(df["index"], df["V Kohli"])
plt.plot(df["index"], df["RG Sharma"])
```

```python
# Adding title, x_label, y_label
plt.plot(df["index"], df["V Kohli"])
plt.plot(df["index"], df["RG Sharma"])
plt.title("V Kohli VS RG Sharma Career Comparison")
plt.xlabel("Year")
plt.ylabel("Score")
```

```python
# Adding color, linestyle, linewidth, marker
# linestyle --> dashed, dotted, dashdot
plt.plot(df["index"], df["V Kohli"], color="green", linestyle="dashed", linewidth =3,marker ="*")
plt.plot(df["index"], df["RG Sharma"], color="red", linestyle="dashdot", marker ="^")
plt.title("V Kohli VS RG Sharma Career Comparison")
plt.xlabel("Year")
plt.ylabel("Score")
```

```python
# Adding Legend
plt.plot(df["index"], df["V Kohli"], color="green", linestyle="dashed", linewidth =3,marker ="*", label ="V Kohli")
plt.plot(df["index"], df["RG Sharma"], color="red", linestyle="dashdot", marker ="*", label ="RG Sharma")
plt.title("V Kohli VS RG Sharma Career Comparison")
plt.xlabel("Year")
plt.ylabel("Score")
# loc --> 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
plt.legend(loc="best")
```

```python
# 1 million is outlierr
price =[10000, 30000, 80000, 70000, 50000, 10000000]
year =[2008, 2009, 2010, 2011, 2012, 2013]
plt.plot(year, price)
#Same can done for x-axis to see prices within some ranfe of years
plt.ylim(0, 2000000)
plt.grid()
```

## Bar Chart
- Biivariate Analysis (2 cols)
- 1st Catagorical and 2nd Numerical

```python
# simple bar chart
colors =["Red","Green", "Blue", "Black"]
children =[10,20,30,10]
plt.bar(colors, children, color="black")
```

```python
# For Horizontal Bars
plt.barh(colors, children, color="black")
```

```python
df =pd.read_csv("datasets\batsman_season_record.csv")
df
```

```python
# Grouped Bar chart
import numpy as np
plt.bar(np.arange(df.shape[0])-0.2 ,df["2015"], width=0.2, label =2015)
plt.bar(np.arange(df.shape[0]), df["2016"], width=0.2, label ="2016")
plt.bar(np.arange(df.shape[0])+0.2, df["2017"], width=0.2, label ="2017")
# change x tics
plt.xticks(np.arange(df.shape[0]), df["batsman"])
plt.xlabel("Players")
plt.ylabel("Score")
plt.legend()
```

```python
arr =np.arange(df.shape[0])-0.2
plt.bar(arr ,df["2015"], label =2015, width=0.2)
```

```python
# Categories Overlapping Problem
values =[10, 20, 10, 15, 16]
colors =["Light red red", "Pink Pink Green", "pink pink pink", "white brown", "reddish blue"]
plt.xticks(rotation ="vertical")
plt.bar(colors, values)
```

```python
df =pd.read_csv("datasets/batsman_season_record.csv")
```

```python
df
```

```python
## Stacked Bar chart
plt.bar(df["batsman"], df["2015"], label ="2015")
plt.bar(df["batsman"], df["2016"], bottom=df["2015"], label ="2016")
plt.bar(df["batsman"], df["2017"], bottom=df["2015"] + df["2016"], label ="2017")
plt.legend()
```

## Subplots

```python
df =pd.read_csv(r"datasets\batter.csv")
df
```

```python
#fig is whole figure
#ax is subplots
# use ax[0, 0] in 2d case
fig, ax =plt.subplots(nrows =2, ncols =1, sharex=True, figsize =(10, 5))
ax[0].scatter(df["avg"], df["strike_rate"] ,color ="blue")
ax[0].set_title("Avg Vs Strike Rate")
ax[0].set_ylabel("Strike Rate")
ax[1].scatter(df["avg"], df["runs"], color="red")
ax[1].set_title("Avg Vs Runs")
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("Runs")

```

## HeatMap

```python
delivery =pd.read_csv("datasets\deliveries.csv")
delivery
```

```python
# Find In All ipl matches how many sixes occures on each ball of each over
temp =delivery[(delivery["batsman_runs"]==6 )& (delivery["ball"].isin([1,2,3,4,5,6]))]
```

```python
grid =temp.pivot_table(index ="over", columns ="ball", values="batsman_runs", aggfunc ="count")
```

```python
grid
```

```python
plt.imshow(grid)
plt.colorbar()
```
