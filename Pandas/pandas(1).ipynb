## Handling Missing Values

```python
import numpy as np
scores =pd.Series([0, 1, np.nan , 2, 3, 4, 5, np.nan])
```

```python
scores
```

```python
# isnull
scores.isnull()
```

```python
# Count null values
scores.isnull().sum()
```

```python
# dropna -->First way to handle missing val is to drop record that have na values
scores.dropna()
```

```python
# fillna --> Second is to fill those value, mostly it is filled by mean
scores.fillna(0)
```

```python
scores
```

```python
# Return where score is 1, 3
# return where score is 1 or 4
(scores==1.0) | (scores==4.0)
```

```python
scores[(scores==1.0) | (scores==4.0)]
```

```python
# but if there is too much conditions 
scores[scores.isin([1.0,2.0,3.0,4.0])]
```
