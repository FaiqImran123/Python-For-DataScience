## GroupBy in Pandas

```python
import pandas as pd
```

```python
movies =pd.read_csv("datasets\imdb-top-1000.csv")
movies
```

```python
# creating groupby object
genre =movies.groupby("Genre")
```

## Applying Builtin aggregate functions on GroupBy

```python
genre.sum()
```

```python
genre.max()
```

```python
genre.min()
```

```python
# Find top 3 genres based on earning
movies.groupby("Genre").sum()["Gross"].sort_values(ascending=False)[0:3]
```

```python
# find genre with highest avg imbd rating
movies[["IMDB_Rating", "Genre"]].groupby("Genre").mean()["IMDB_Rating"].sort_values(ascending=False).index[0]
```

```python
# Find director with most popularity
movies.groupby("Director").sum()["No_of_Votes"].sort_values(ascending=False).index[0]
```

## Group By Attributes and Methods

```python
# finding length of total groups
len(movies.groupby("Genre"))
```

```python
# size()
movies.groupby("Genre").size()
```

```python
# first movie of each genre/ same for last
movies.groupby("Genre").first()
```

```python
# groups --> Index-->Genre, Value-->Index postion of all action movies
genre.groups
```

```python
genre.describe()
```

```python
# let you want to apply differ agg function on differ cols
# example totoal votes and Max rating
genre.agg({"No_of_Votes": "sum", "IMDB_Rating": "max"})
```

```python
# Looping on groups
for group, data in genre:
    print(group)

    
```

```python
# Highest rated movies of each genre
def highest_rated(data):
    return data[data["IMDB_Rating"]==data["IMDB_Rating"].max()]
```

```python
genre.apply(highest_rated)
```

```python
# Number of movies starting with A for each group
def find(data):
    return data["Series_Title"].apply(lambda x : x.startswith("A")).sum()
```

```python
genre.apply(find)
```

```python
# Ranking of each movie in group according to IMDB_Score
def rating(data):
    data["ranking"] = data["IMDB_Rating"].rank(ascending =False)
    return data
```

```python
genre.apply(rating)
```

```python
# Groupby on multiple cols
movies.groupby(["Director", "Star1"])
```

```python
# most earning (director + star1 combo)
movies.groupby(["Director", "Star1"]).sum()["Gross"].sort_values(ascending=False)
```

```python
# find the best(in terms of avg meta score) actor + director combo
movies.groupby(["Director", "Star1"])["Metascore"].mean().sort_values(ascending=False)
```

```python
deliveries =pd.read_csv("datasets\deliveries.csv")
deliveries
```

```python
# top 10 batsman in terms of run
deliveries.groupby("batsman").sum()["batsman_runs"].sort_values(ascending=False)
```

```python
# Batsman with max number of sixes
deliveries[deliveries["batsman_runs"]==6].groupby("batsman")["batsman"].count().sort_values(ascending=False)
```

```python
# create a function that return highest of anyy batsman
def highest(batsman, df):
    return df[df["batsman"]==batsman].groupby("match_id").sum()["batsman_runs"].sort_values(ascending =False).values[0]
```

```python
highest("CH Gayle", deliveries)
```
